```python
import matplotlib.pyplot as plt

# Data
block_sizes = [512, 1024, 2048, 3072, 4096, 7680, 15360]
sign = [9.86, 9.84, 9.94, 9.93, 9.98, 9.93, 10.23]
verify = [9.92, 9.88, 9.83, 9.69, 9.95, 9.88, 9.91]
encrypt = [9.9, 9.95, 9.86, 9.9, 8.89, 9.97, 9.79]
decrypt = [9.91, 9.93, 9.9, 9.96, 8.73, 9.96, 10.22]

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(block_sizes, sign, marker='o', label='Sign')
plt.plot(block_sizes, verify, marker='o', label='Verify')
plt.plot(block_sizes, encrypt, marker='o', label='Encrypt')
plt.plot(block_sizes, decrypt, marker='o', label='Decrypt')

plt.title('Performance vs. Block Size')
plt.xlabel('Block Size')
plt.ylabel('Time (seconds)')
plt.xticks(block_sizes)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
